# utils/lib/

Questa cartella contiene le cose importanti per il progetto (non sono da modificare direttamente).

- `chiral_potential.py`: Implementa la teoria EFT chirale. Calcola i termini del potenziale $V$ (LO, NLO) basandosi sulle costanti LEC.
- `nn_studio.py`: È l'analizzatore principale. Gestisce la cinematica (trasformazione laboratorio-centro di massa), la discretizzazione della mesh di Gauss-Legendre e il calcolo degli sfasamenti ($\delta$).
- `lec_values.py`: Contiene i valori numerici predefiniti delle costanti di accoppiamento per i vari ordini.
- `granada_phases.py`: Carica i dati sperimentali di riferimento (i "punti neri" nei grafici delle slide) per il confronto.
- `auxiliary.py`: Contiene funzioni di supporto, molto probabilmente inclusi i calcoli geometrici per il momento di quadrupolo.


# utils/esempi/

Qui ci sono i file che dovrai eseguire e completare:

1. Esplorazione: `plot_potential.py` e `plot_phase_shift.py`
- eseguire e visualizzare i grafici per capire come il potenziale e gli sfasamenti cambiano con l'ordine di chiralità (LO vs NLO) e con i dati sperimentali.

2. Il cuore del Punto 2: `simulate_deuteron.py` 
- Logica: Il codice usa `nn.setup_Vmtx` per ottenere il potenziale nel canale $^3S_1$-$^3D_1$ , costruisce l'Hamiltoniana $H=T+V$ discretizzata e la diagonalizza per trovare l'energia di legame $E_0$.  
- Estrarre l'energia (circa -2.22 MeV) e usare gli autovettori per calcolare il momento di quadrupolo.

3. Analisi Bayesiana (Punti 3, 4 e 5)
- Definire i Priori: Usare distribuzioni gaussiane per le LEC (es. $C_{1S0}$) centrate in 0.  
- Likelihood: Creare una funzione che calcoli quanto le tue predizioni (da `nn_studio` e `simulate_deuteron`) si discostino dai dati sperimentali.  
- Campionamento: Generare distribuzioni come quelle mostrate nell'Appendice B, dove si vede il "cammino" dell'algoritmo verso i valori più probabili delle LEC.
